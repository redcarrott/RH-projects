## deeper dive into Definitive Healthcare's IDN Name column 

```python
#idn_names
```

```python
df_IDNs[df_IDNs['IDN Name'].str.contains('Atrium Health')]
```

```python
idn_list
```

```python
new_list = []
for name in idn_list:
    if 'FKA' in name:
        split_name = name.split(' (FKA ')
        new_name_list = [x.replace(")","") for x in split_name]
        for new_name in new_name_list:
            new_list.append(new_name)
            continue
        continue
    if 'AKA' in name:
        split_name = name.split(' (AKA ')
        new_name_list = [x.replace(")","") for x in split_name]
        for new_name in new_name_list:
            new_list.append(new_name)
            continue
        continue
    if 'FKA' or 'AKA' not in name:
        new_list.append(name)
```

```python
#new_list
```

```python
# Column contains particular value by multiple strings.
idn_names = '|'.join(new_list)
contacts_IDN_match = df[df['Company'].str.contains(idn_names)]
contacts_IDN_match
```

```python
# export as csv
contacts_IDN_match.to_csv('contacts_IDN_match.csv')
```
